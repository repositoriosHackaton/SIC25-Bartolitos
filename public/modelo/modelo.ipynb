<h2>Primero deremos cargar nuestro dataset</h2>

```python
import pandas as pd

dataset='weather_classification_data.csv'

weather_df=pd.read_csv(dataset)

weather_df.head(10)
```

```python
print(weather_df['Cloud Cover'].unique())
```

<h3>Validamos que nuestro dataframe no poseea valores nulos 
</h3>

```python
weather_df.isnull().sum()/len(weather_df)*100
```

<h3>Limpiamos los Datos</h3>

```python
print(weather_df['Cloud Cover'].unique())
weather_df['Cloud Cover'] = weather_df['Cloud Cover'].replace({'partly cloudy': 0, 'clear': 1, 'overcast': 2, 'cloudy': 3})
print(weather_df['Cloud Cover'].unique())
```

```python
print(weather_df['Season'].unique())
weather_df['Season'] = weather_df['Season'].replace({'Winter': 0, 'Spring': 1, 'Summer': 2, 'Autumn': 3})
print(weather_df['Season'].unique())
```

```python
print(weather_df['Location'].unique())
weather_df['Location'] = weather_df['Location'].replace({'inland': 0, 'mountain': 1, 'coastal': 2})
print(weather_df['Location'].unique())
```

```python
print(weather_df['Weather Type'].unique())
weather_df['Weather Type'] = weather_df['Weather Type'].replace({'Rainy': 0, 'Cloudy': 1, 'Sunny': 2, 'Snowy': 3})
```

```python
print(weather_df['Weather Type'].unique())
```

```python
Wy=weather_df['Weather Type']
Wx=weather_df.drop(columns=['Weather Type'])
```

```python
print(Wy.unique())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Wx_scaled= scaler.fit_transform(Wx)
Wx_scaled=pd.DataFrame(Wx_scaled, columns=Wx.columns)
Wx_scaled.head()
```

```python
import joblib

# Guardar el scaler en un archivo
joblib.dump(scaler, "scaler.pkl")
```

```python
from sklearn.model_selection import train_test_split
Wx_train, Wx_test, Wy_train, Wy_test = train_test_split(Wx_scaled, Wy, test_size=0.7, random_state=532004)
```

```python
from sklearn.model_selection import train_test_split
Wx_train, Wx_test, Wy_train, Wy_test = train_test_split(Wx, Wy, test_size=0.7, random_state=532004)
```

```python
distribucion = Wy_train.value_counts()
print("Distribución de elementos únicos:")
print(distribucion)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {'weather': ['Rainy']*1008 + ['Cloudy']*1006 + ['Snowy']*1001 + ['Sunny']*945}
df = pd.DataFrame(data)

# Obtener la distribución de elementos únicos
distribucion = df['weather'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
distribucion.plot(kind='bar', color=['blue', 'gray', 'green', 'yellow'])

# Añadir título y etiquetas
plt.title('Distribución del clima')
plt.xlabel('Tipo de clima')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()

#25% en todos menos soleado que tuvo un 23
```

```python
from sklearn.ensemble import RandomForestClassifier

RF_model=RandomForestClassifier(n_estimators=100, random_state=532004, min_samples_leaf=8)
RF_model.fit(Wx_train, Wy_train) 
```

```python
RF_model.score(Wx_test, Wy_test)
```

```python
Wy_pred = RF_model.predict(Wx_test)

Wy_pred
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np


cm = confusion_matrix(Wy_test, Wy_pred)

```
